```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month" , columns="year" , values="passengers")
flight_df.head()
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(flight_df)
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(flight_df , annot = True)
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(flight_df , annot = True , fmt = "d")
```
